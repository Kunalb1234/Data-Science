```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Dataset .csv")
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
#TASK 1
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

features = data[['Restaurant ID', 'Country Code', 'Longitude', 'Latitude', 'Average Cost for two', 'Price range', 'Votes']]
target = data['Aggregate rating']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

```

```python
#linear regression model
model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print('\nLinear regression:')
print(f'Mean Squared Error (MSE): {mse:.4f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
print(f'R-squared (R2): {r2:.4f}')
```

```python
# Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor

decision_tree_model = DecisionTreeRegressor(random_state=42)
decision_tree_model.fit(X_train, y_train)

y_pred_decision_tree = decision_tree_model.predict(X_test)

mse_decision_tree = mean_squared_error(y_test, y_pred_decision_tree)
rmse_decision_tree = np.sqrt(mse_decision_tree)
r2_decision_tree = r2_score(y_test, y_pred_decision_tree)

print('\nDecision Tree Regression:')
print(f'Mean Squared Error (MSE): {mse_decision_tree:.4f}')
print(f'Root Mean Squared Error (RMSE): {rmse_decision_tree:.4f}')
print(f'R-squared (R2): {r2_decision_tree:.4f}')
```

```python
# Random Forest Regression
from sklearn.ensemble import RandomForestRegressor

random_forest_model = RandomForestRegressor(random_state=42)
random_forest_model.fit(X_train, y_train)

y_pred_random_forest = random_forest_model.predict(X_test)

mse_random_forest = mean_squared_error(y_test, y_pred_random_forest)
rmse_random_forest = np.sqrt(mse_random_forest)
r2_random_forest = r2_score(y_test, y_pred_random_forest)

print('\nRandom Forest Regression:')
print(f'Mean Squared Error (MSE): {mse_random_forest:.4f}')
print(f'Root Mean Squared Error (RMSE): {rmse_random_forest:.4f}')
print(f'R-squared (R2): {r2_random_forest:.4f}')
```

```python
#By comparing Linear Regression,Decision Tree Regression and Random Forest Regression I conclude that Random Forest Regression is the best fit.
#Because Random Forest Regression has the Lowest MSE and RMSE.
#It also has higgest R-squared value.
```
