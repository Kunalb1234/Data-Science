```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Dataset .csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
#Task 1
# Percentage of restaurants that offers
data['Has Table booking'] = data['Has Table booking'].map({'Yes': True, 'No': False})
data['Has Online delivery'] = data['Has Online delivery'].map({'Yes': True, 'No': False})
```

```python
data.head()
```

```python
data['Has Table booking'].value_counts()
```

```python
# Table booking
table_booking_percentage = (data['Has Table booking'].sum() / len(data)) * 100

# Online delivery
online_delivery_percentage = (data['Has Online delivery'].sum() / len(data)) * 100


print(f"Percentage of Restaurants with Table Booking: {table_booking_percentage:.2f}%")
print(f"Percentage of Restaurants with Online Delivery: {online_delivery_percentage:.2f}%")
```

```python
data.head()
```

```python
#AVERAGE RATINGS OF RESTAURANTS WITH AND WITHOUNT TABLE BOOKING
print("Total number of rows in the dataset:", len(data))
print("Count of restaurants with table booking:", data[data['Has Table booking']].shape[0])
```

```python
count_with_booking = data[data['Has Table booking']].shape[0]

if count_with_booking > 0:
    # Calculate the average ratings for restaurants with and without table booking
    average_rating_with_booking = data[data['Has Table booking']]['Aggregate rating'].mean()
    average_rating_without_booking = data[~data['Has Table booking']]['Aggregate rating'].mean()
    
    print(f"Average Rating with Table Booking: {average_rating_with_booking:.2f}")
    print(f"Average Rating without Table Booking: {average_rating_without_booking:.2f}")
else:
    print("No restaurants with table booking in the dataset.")
```

```python
data.head()
```

```python
#Analyzing the availability of online delivery among restaurants with different price ranges.
delivery_by_price = data.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack()

import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=delivery_by_price.index, y=delivery_by_price[True], color='green', label='With Online Delivery')
sns.barplot(x=delivery_by_price.index, y=delivery_by_price[False], color='red', bottom=delivery_by_price[True], label='Without Online Delivery')

plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.title('Availability of Online Delivery by Price Range')

plt.legend(title='Online Delivery', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Task 2
#The most common price range among all the restaurants
most_common_price_range = data['Price range'].mode().values[0]
print(f"The most common price range among all the restaurants is: {most_common_price_range}")
```

```python
# Calculate the average rating for each price range
average_rating_by_price = data.groupby('Price range')['Aggregate rating'].mean()

print("Average Rating by Price Range:")
print(average_rating_by_price)
```

```python
# Identifying the color that represents the highest average rating

avg_rating_by_price = data.groupby('Price range')['Aggregate rating'].mean().sort_values(ascending=False)

highest_avg_rating_color = data[data['Price range'] == avg_rating_by_price.idxmax()]['Rating color'].iloc[0]

print(f"Highest Average Rating Color: {highest_avg_rating_color}")
print(f"Highest Average Rating Price Range: {avg_rating_by_price.idxmax()}")
```

```python
# Counting the number of services a restaurant offers
data['Services Count'] = data[['Has Table booking', 'Has Online delivery']].sum(axis=1)

print(data[['Has Table booking', 'Has Online delivery', 'Services Count']])
```

```python
#This is the bar plot representaion of the Service Count
import matplotlib.pyplot as plt

data['Services Count'] = data[['Has Table booking', 'Has Online delivery']].sum(axis=1)

data['Services Count'].value_counts().sort_index().plot(kind='bar')

plt.title('Distribution of Services Count')
plt.xlabel('Number of Services')
plt.ylabel('Count')
plt.show()
```
